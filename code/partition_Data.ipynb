```python
# Load libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load US Census Demographic Data
census_data = pd.read_csv('US_census_data_2017.csv')
census_data.head()
```

```python
#Divide into train and test datasets
msk = np.random.rand(len(census_data)) <= 0.75
train = census_data[msk]
test = census_data[~msk]
```

```python
#Determine Length of train and test datasets
print(len(test))
print(len(train))
```

```python
census_train = pd.DataFrame(data=train)
census_train.to_csv("census_train.csv")
census_train.head()
```

```python
census_test = pd.DataFrame(data=test)
census_test.to_csv("census_test.csv")
census_test.head()
```
